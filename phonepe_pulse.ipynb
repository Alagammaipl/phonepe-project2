```python
#Once created the clone of GIT-HUB repository then,
#Required libraries for the program

import pandas as pd
import json
import os

#This is to direct the path to get the data as states

path="/content/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}
for i in Agg_state_list:
    a=path+i+"/"
    Agg_yr=os.listdir(a)    
    for j in Agg_yr:
        b=a+j+"/"
        Agg_yr_list=os.listdir(b)        
        for k in Agg_yr_list:
            c=b+k
            Data=open(c,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transaction_type'].append(Name)
              clm['Transaction_count'].append(count)
              clm['Transaction_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
aggregated_transactions = pd.DataFrame(clm)
aggregated_transactions
```

```python
%%writefile app.py

#importing needed modules

import streamlit as st
import numpy as np
import pandas as pd
import json
import sqlite3
import plotly.express as px

#establishing sql connection
conn = sqlite3.connect('my_database.db')

#extracting data from mysql database

aggregated_transactions = pd.read_sql_query("select * FROM aggregated_transactions", conn)
top_transactions_dis = pd.read_sql_query("select * FROM top_transactions_dis", conn)
top_transactions_pincodes = pd.read_sql_query("select * FROM top_transactions_pincodes", conn)
map_transactions = pd.read_sql_query("select * FROM map_transactions", conn)
map_users = pd.read_sql_query("select * FROM map_users", conn)
top_users_dis = pd.read_sql_query("select * FROM top_users_dis", conn)
top_users_pin = pd.read_sql_query("select * FROM top_users_pin", conn)
agg_users = pd.read_sql_query("select * FROM agg_users", conn)
agg_usersbydevice = pd.read_sql_query("select * FROM agg_usersbydevice", conn)
conn.commit()

#creating heading

st.title("**_PhonePe Pulse Data Visualization_**")
st.caption("This app is an user-interactive data Visualization")



#creating sidebar with selectboxes to filter data acordingly

st.sidebar.title("Filters")
x = st.sidebar.selectbox("All India or Statewise", ("All india", "Statewise"))
a = st.sidebar.selectbox('Select transactions or users', ('Transactions', 'Users'))
y = st.sidebar.selectbox("Data quaterwise or overall", ("Overall", "quaterwise"))

# all india Transactions


if x == "All india":
 if a == "Transactions":
   if y == "quaterwise":
     b = st.sidebar.selectbox('year', ('2018', '2019', '2020', '2021', '2022'))
     c = st.sidebar.selectbox('Quartile', ('1', '2', '3', '4'))
     col1,col2,col3 = st.columns(3)
     aggtran =aggregated_transactions[aggregated_transactions['Year'] == b]
     aggtrans = aggtran[aggtran['Quater'] == int(c)]
     with col1:
       st.header("**Transactions**")
       st.write("Total transactions", aggtrans.Transaction_count.sum())
       st.write("Total payment", aggtrans.Transaction_amount.sum())
       st.write("Avg payment per transaction",aggtrans.Transaction_amount.sum()/aggtrans.Transaction_count.sum())
     e = aggtrans.Transaction_type.unique()
     with col2:
       st.write()
     with col3:
       st.header("_Categories_")
       for i in range (len(e)):
         u = aggtrans[aggtrans['Transaction_type'] == e[i]]
         st.write(e[i], u.Transaction_count.sum())
    
     st.write("**Geo-visualization**")     
     fig = px.choropleth(aggtrans,
                        geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
                        featureidkey='properties.ST_NM',
                        locations='State',color='Transaction_amount',hover_data =['Transaction_count', 'State'],projection="mercator",
                        color_continuous_scale='plotly3')
     fig.update_geos(fitbounds="locations", visible = False)
     st.plotly_chart(fig, theme = "streamlit")
   else:
      col1,col2,col3 = st.columns(3)
      with col1:
       st.header("**Transactions**")
       st.write("Total transactions", aggregated_transactions.Transaction_count.sum())
       st.write("Total payment", aggregated_transactions.Transaction_amount.sum())
       st.write("Avg payment per transaction",aggregated_transactions.Transaction_amount.sum()/aggregated_transactions.Transaction_count.sum())
       e = aggregated_transactions.Transaction_type.unique()
      with col2:
       st.write()
      with col3:
       st.header("_Categories_")
       for i in range (len(e)):
         u = aggregated_transactions[aggregated_transactions['Transaction_type'] == e[i]]
         st.write(e[i], u.Transaction_count.sum())



# users all india




 else:
   if y == "quaterwise":
     b = st.sidebar.selectbox('year', ('2018', '2019', '2020', '2021', '2022'))
     c = st.sidebar.selectbox('Quartile', ('1', '2', '3', '4'))
     agguse =agg_users[agg_users['Year'] == b]
     aggus = agguse[agguse['Quater'] == int(c)]
     st.write("Total no of app opens:",aggus.appOpens.sum())
     st.write("Total newly registered users", aggus.registered_users.sum())
     fig = px.choropleth(aggus,
                        geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
                        featureidkey='properties.ST_NM',
                        locations='State',color='registered_users',hover_data =['appOpens', 'State'],projection="mercator",
                        color_continuous_scale='plotly3')
     fig.update_geos(fitbounds="locations", visible = False)
     st.plotly_chart(fig, theme = "streamlit")
   else:
     st.write("Total no of app opens:",agg_users.appOpens.sum())
     st.write("Total newly registered users", agg_users.registered_users.sum())




   
# statewisetransactions



if x == "Statewise":
  f = st.sidebar.selectbox("Select a state", ['Jharkhand', 'Tamil Nadu', 'West Bengal', 'Nagaland', 'Gujarat',
       'Telangana', 'Karnataka', 'Sikkim', 'Goa', 'Delhi', 'Haryana',
       'Ladakh', 'Odisha', 'Lakshadweep', 'Maharashtra',
       'Dadra and Nagar Haveli and Daman and Diu', 'Punjab', 'Puducherry',
       'Tripura', 'Arunachal Pradesh', 'Manipur',
       'Andaman & Nicobar Island', 'Himachal Pradesh', 'Madhya Pradesh',
       'Mizoram', 'Kerala', 'Chhattisgarh', 'Chandigarh',
       'Jammu & Kashmir', 'Meghalaya', 'Uttarakhand', 'Rajasthan',
       'Bihar', 'Andhra Pradesh', 'Assam', 'Uttar Pradesh'])

  if a == "Transactions":
    if y == "quaterwise":
      b = st.sidebar.selectbox('year', ('2018', '2019', '2020', '2021', '2022'))
      c = st.sidebar.selectbox('Quartile', ('1', '2', '3', '4'))
      tab1, tab2 = st.tabs(["overall", "districts"])
      with tab1:
       col1,col2,col3 = st.columns(3)
       aggtran =aggregated_transactions[aggregated_transactions['Year'] == b]
       aggtrans = aggtran[aggtran['Quater'] == int(c)]
       aggst = aggtrans[aggtrans['State'] == f]

       toptran = top_transactions_dis [top_transactions_dis['Year'] == b]
       toptrans = toptran[toptran['Quater'] == int(c)]
       topst = toptrans[toptrans['State'] == f]
       toptra = topst.drop(['Year', 'Quater', 'State'], axis = 1)
      
       toptranp = top_transactions_pincodes [top_transactions_pincodes['Year'] == b]
       toptransp = toptranp[toptranp['Quater'] == int(c)]
       topstp = toptransp[toptransp['State'] == f]
       toptrap = topstp.drop(['Year', 'Quater', 'State'], axis = 1)
      
       maptrans = map_transactions[map_transactions['Year'] == b]
       maptran = maptrans[maptrans['Quater'] == int(c)]
       mapst = maptran[maptran['State'] == f]
       maptra = mapst.drop(['Year', 'Quater', 'State'], axis = 1)
       with col1:
        st.header("**Transactions**")
        st.write("Total transactions", aggst.Transaction_count.sum())
        st.write("Total payment", aggst.Transaction_amount.sum())
        st.write("Avg payment per transaction",aggst.Transaction_amount.sum()/aggtrans.Transaction_count.sum())
       with col2:
        st.write()
       with col3:
        st.header("_Categories_")
        e = aggtrans.Transaction_type.unique()
        for i in range (len(e)):
           u = aggtrans[aggtrans['Transaction_type'] == e[i]]
           st.write(e[i], u.Transaction_count.sum())
       col1, col2 = st.columns(2)
       with col1:
         st.write("_Top 10 Districts_")
         st.write(toptra)
       with col2:
         st.write("_Top 10 Pincodes_")
         st.write(toptrap)
      with tab2:
        st.write(maptra)
    else:
      col1,col2,col3 = st.columns(3)
      with col1:
       aggregated_transactions = aggregated_transactions[aggregated_transactions['State'] == f]
       st.header("**Transactions**")
       st.write("Total transactions", aggregated_transactions.Transaction_count.sum())
       st.write("Total payment", aggregated_transactions.Transaction_amount.sum())
       st.write("Avg payment per transaction", aggregated_transactions.Transaction_amount.sum()/aggregated_transactions.Transaction_count.sum())
      with col2:
        st.write()
      with col3:
        st.header("_Categories_")
        e = aggregated_transactions.Transaction_type.unique()
        for i in range (len(e)):
          u = aggregated_transactions[aggregated_transactions['Transaction_type'] == e[i]]
          st.write(e[i], u.Transaction_count.sum())



#Statewise users


  else:
    if y == "quaterwise":
      b = st.sidebar.selectbox('year', ('2018', '2019', '2020', '2021', '2022'))
      c = st.sidebar.selectbox('Quartile', ('1', '2', '3', '4'))
      agguse =agg_users[agg_users['Year'] == b]
      aggus = agguse[agguse['Quater'] == int(c)]
      aggstus = aggus[aggus['State'] == f]

      topuse =top_users_dis[top_users_dis['Year'] == b]
      topus = topuse[topuse['Quater'] == int(c)]
      topstus = topus[topus['State'] == f]
      topstus = topstus.drop(['Year', 'State', 'Quater'], axis = 1)

      topusep =top_users_pin[top_users_pin['Year'] == b]
      topusp = topusep[topusep['Quater'] == int(c)]
      topstusp = topusp[topusp['State'] == f]
      topstusp = topstusp.drop(['Year', 'State', 'Quater'], axis = 1)

      mapuse =map_users[map_users['Year'] == b]
      mapus = mapuse[mapuse['Quater'] == int(c)]
      mapstus = mapus[mapus['State'] == f]
      mapstus = mapstus.drop(['Year', 'State', 'Quater'], axis = 1)
      tab1,tab2 = st.tabs(["overall", "districts"])

      with tab1:
       st.write("Total no of app opens",aggstus.appOpens.sum())
       st.write("Total newly registered users", aggstus.registered_users.sum())
       col1,col2 = st.columns(2)
       with col1:
        st.write("_Top 10 Districts_")
        st.write(topstus)
       with col2:
        st.write("_Top 10 pincodes_")
        st.write(topstusp)
      with tab2:
       st.write(mapstus)
    else:
      aggst = agg_users[agg_users['State'] == f]
      st.write("Total no of app opens till 2022",aggst.appOpens.sum())
      st.write("Total registered users till 2022", aggst.registered_users.sum())



  
```
